```python
# Plot hist of total_paid
# orders["total_paid"].plot(kind="bar")
```

```python
# Plot state of orders and order_id
# order_id are unique
orders["order_id"].groupby(orders["state"]).count().plot(kind="barh")
```

```python
#plot order month and number of orders

ord_short_dates = orders["created_date"].dt.to_period('M')
ord_id_count = orders.groupby(ord_short_dates)["order_id"].count()
ord_id_count.plot(kind="barh")
```
